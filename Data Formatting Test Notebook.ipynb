```python
import pandas as pd
import numpy as np
import random
import names
```

```python
x = np.random.randint(low=0, high=100, size=300)
y = np.random.randint(low=0, high=100, size=300)

types = [80,60,34,49,3,5,9,6]
p_dist = []
for i in types:
    p_dist.append(i/sum(types))

from numpy.random import choice
draw = choice(['a','b','c','d','e','f','g','h'],300,
              p= p_dist)

raw_data = pd.DataFrame({'x':x,'y':y,'type':draw})
```

```python
print(raw_data['type'].value_counts())

sizes = list(raw_data['type'].value_counts())
labels = list(raw_data['type'].value_counts().index)
```

```python
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

```python
def dataGrouping(df,groups=4):
    sizes = list(df['type'].value_counts())
    labels = list(df['type'].value_counts().index)    
    new_df = df.copy()
    
    for i in labels[groups:]:
        new_df['type'] = new_df['type'].replace(i,'other')
    return new_df
```

```python
new_df = dataGrouping(raw_data,3)
new_df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt
new_sizes = list(new_df['type'].value_counts())
new_labels = list(new_df['type'].value_counts().index)

import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots()
ax1.pie(new_sizes,labels=new_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
